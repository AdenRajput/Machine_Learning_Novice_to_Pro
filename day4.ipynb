```python
# columns of the data
chotu_data.columns
```

```python
chotu_data.info()
```

```python
# stat for the data
chotu_data.describe() # only numerical data
chotu_data.describe(include='all')
```

```python
# reindexing of chotu data
new_index = range(1,11)
chotu_data = chotu_data.reset_index(drop=True).reindex(new_index)
```

```python
chotu_data.head()
```
