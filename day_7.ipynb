# Data Visualization on Tips Data set using Matplotlib and Seaborn:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
# Q1: What is the distribution of the total bill amount?
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title("Distribution of the Total bill")
plt.xlabel("Amount of Total Bill")
plt.ylabel("Frequency")
plt.show()
```

- Total bill is showing skewed result which implies that on average the total amount of the bill falls between 10USD to 20USD. Thus, it seems to be a good option to dine-in.

```python
#3. On which day the resturant is open..!!
print(tips.day.unique())
sns.countplot(x='day',data=tips)
```

```python
#4. Which day of the week has the highest total bill amounts on average..!!!
sns.barplot(x='day',y='total_bill',data=tips, ci=None)
plt.title("Average total bill by day")
plt.xlabel("Days of the Week")
plt.ylabel("Average total bill")
plt.show()
```

```python
#5. How does the tip amount vary by time of the meal.

sns.boxplot(data=tips, x='time',y='tip')
plt.title("Amount of the TIP by Time")
plt.xlabel("time")
plt.ylabel("Tip Amount")
plt.show()
```
