```python
#loading the libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line plot in Matplotlib:

```python
# data for the plot
x = [1,2,3,4,5]
y = [2,14,6,18,10]

#plotting line plot
plt.plot(x,y)
```

```python
## refining the initial plot

# data for the plot
x = [1,2,3,4,5]
y = [2,14,6,18,10]

#plotting line plot
plt.plot(x,y)
plt.title('Sada sa line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
## mazeed refinement
## refining the initial plot

# data for the plot
x = [1,2,3,4,5]
y = [2,14,6,18,10]

#plotting line plot
plt.plot(x,y,color='purple',linestyle = '--', label = 'GARMI')
plt.legend()
plt.title('Sada sa line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

### Scatter plot in Matplotlib:

```python
#scatter plot
plt.scatter(x,y,color='red',marker='x')
plt.title("x ka scatter plot")
plt.xlabel("yahan x ki value hay")
plt.ylabel("yahan y ki value hay")
plt.show()
```

### Multiple plots using matplotlib:

```python
plt.figure(figsize=(10,4)) #canvas ka size bta rhay hayn 

plt.subplot(2,1,1) # 2 rows, 1 column, abhi hum 1st graph btanay waly hayn
plt.plot(x,y)
plt.title("Pehla plot")

plt.subplot(2,1,2) # 2 rows, 1 column, abhi hum 1st graph btanay waly hayn
plt.scatter(x,y)
plt.title("Dusra plot")

plt.tight_layout() #overlapping kthm hojati hay
plt.show()
```

```python
plt.figure(figsize=(10,4)) #canvas ka size bta rhay hayn 

plt.subplot(1,2,1) # 2 rows, 1 column, abhi hum 1st graph btanay waly hayn
plt.plot(x,y)
plt.title("Pehla plot")

plt.subplot(1,2,2) # 2 rows, 1 column, abhi hum 1st graph btanay waly hayn
plt.scatter(x,y)
plt.title("Dusra plot")

plt.tight_layout() #overlapping kthm hojati hay
plt.show()
```

### Bar plot in Matplotlib

```python
# sample data
khanay = ['alu gobi','pasta','pizza','bhindi']
tadad = [2, 14, 1, 5]

#creating a bar plot
plt.bar(khanay,tadad)

#adding title and labels
plt.xlabel("KHANAY")
plt.ylabel("No. of plates")
plt.title("RAAT KA KHANA")

plt.show()
```

## Seaborn Barplot:

```python
# tips dataset
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.barplot(y='tip',x='day',data=tips)
plt.title("Colorful Din")
plt.show()
```

```python
# seaborn scatter plot
sns.scatterplot(x='tip',y='total_bill',data=tips)
```

```python
# count plot
## kitnay adami thay...
print(tips['sex'].value_counts())
sns.countplot(x='sex',data=tips)

sns.despine() # upr or neechay line khtm 
plt.show()
```

```python
# count plot
## kitnay adami thay...
print(tips['sex'].value_counts())
sns.countplot(x='sex',data=tips,palette='dark', hue='smoker')

sns.despine() # upr or neechay line khtm 
plt.show()
```

```python
sns.catplot(x='sex',hue = 'smoker', col = 'day', data = tips, kind = 'count')
```

```python

#sns.set_theme(style="whitegrid")


# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker",
               split=True, inner="quart", linewidth=1,
              palette={"Yes": "b", "No": ".85"})
sns.despine(left=True)
```
